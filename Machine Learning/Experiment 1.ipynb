### Step 1: Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Generate Random Dataset

```python
np.random.seed(42)

data = {
    'Age': np.random.randint(18, 60, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Salary': np.random.randint(30000, 150000, 100)
}

df = pd.DataFrame(data)

# Display the first few rows
df.head()
```

### Step 3: Create a Histogram of Age

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Step 4: Create a Pie Chart for Gender Distribution

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Gender Distribution', fontsize=16)
plt.show()
```

### Step 5: Create a Scatter Plot of Age vs Salary

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Salary'], alpha=0.7, color='green')
plt.title('Age vs Salary', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

### Step 6: Create a Bar Chart for Gender Count and Create a Heatmap of the Dataset

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Gender Count', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Compute correlation matrix using only numeric columns
corr_matrix = df[['Age', 'Salary']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
